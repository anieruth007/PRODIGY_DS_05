```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('RTA Dataset.csv')
print(df)
```

```python
df.info()
df.isnull().sum()
```

```python
sns.countplot(x='Sex_of_casualty', data=df)
plt.title('Distribution of Casualties by Sex')
plt.show()
```

```python
sns.countplot(x='Age_band_of_casualty', data=df)
plt.title('Distribution of Casualties by Age Band')
plt.show()
```

```python
sns.countplot(x='Casualty_severity', data=df.dropna())
plt.title('Distribution of Casualties by Severity')
plt.show()
```

```python
sns.countplot(x='Accident_severity', data=df.dropna())
plt.title('Distribution of Accidents by Severity')
plt.show()
```

```python
sns.countplot(x='Cause_of_accident', data=df.dropna())
plt.title('Distribution of Accidents by Cause')
plt.show()
```

```python
sns.countplot(x='Vehicle_driver_relation', data=df.dropna())
plt.title('Distribution of Accidents by Vehicle Driver Relation')
plt.show()
```

```python
df['Time'] = pd.to_datetime(df['Time']).dt.hour
sns.countplot(x='Time', data=df)
plt.title('Distribution of Accidents by Time of Day')
plt.show()
```

```python
sns.countplot(x='Vehicle_movement', data=df)
plt.title('Distribution of Accidents by Vehicle Movement')
plt.show()
```
